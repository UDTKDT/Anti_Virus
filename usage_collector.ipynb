```python
from scapy.all import *
import matplotlib.pyplot as plt

# Pcap 파일 경로 설정
pcap_file = 'captured_packets.pcap'

# 패킷 개수를 저장할 리스트 초기화
packet_lengths = []

# Pcap 파일 읽기
packets = rdpcap(pcap_file)

# 각 패킷 길이를 packet_lengths 리스트에 저장
for pkt in packets:
    packet_lengths.append(len(pkt))

# 패킷 개수와 패킷 길이 시각화
plt.figure(figsize=(10, 6))
plt.plot(range(1, len(packet_lengths) + 1), packet_lengths, marker='o', linestyle='-', color='b')
plt.xlabel('Packet Number')
plt.ylabel('Packet Length (bytes)')
plt.title('Packet Lengths in Pcap File')
plt.grid(True)
plt.tight_layout()
plt.show()
```
