```python
import matplotlib.pyplot as plt

# 파일에서 데이터를 읽어오는 함수
def read_data_from_file(filename):
    with open(filename, 'r') as file:
        data = file.read()
    return data

# CPU 사용량 시각화 함수
def visualize_cpu_usage(cpu_data):
    try:
        # 데이터 전처리: 필요한 부분 추출
        lines = cpu_data.splitlines()
        start_line = lines.index('## CPU Usage:') + 1  # 데이터 시작 라인
        cpu_data = lines[start_line:]  # 필요한 데이터 부분만 추출

        # 데이터 파싱
        cpu_cores = []
        cpu_usage = []
        for line in cpu_data:
            if line.startswith('CPU'):
                parts = line.split()
                if len(parts) < 3:
                    continue
                core = parts[1]
                usage = float(parts[-1])
                cpu_cores.append(core)
                cpu_usage.append(usage)

        # 시각화
        plt.figure(figsize=(10, 6))
        plt.bar(cpu_cores, cpu_usage, color='skyblue')
        plt.xlabel('CPU Core')
        plt.ylabel('Usage (%)')
        plt.title('CPU Usage')
        plt.ylim(0, 100)
        plt.grid(True)
        plt.tight_layout()
        plt.show()

    except ValueError as ve:
        print(f"ValueError occurred: {ve}")
    except IndexError as ie:
        print(f"IndexError occurred: {ie}")
    except Exception as e:
        print(f"Exception occurred: {e}")


# 네트워크 사용량 시각화 함수
def visualize_network_usage(network_data):
    try:
        # 데이터 전처리: 필요한 부분 추출
        lines = network_data.splitlines()
        start_line = lines.index('## Network Usage:') + 1  # 데이터 시작 라인
        network_data = lines[start_line:]  # 필요한 데이터 부분만 추출

        # 데이터 파싱 및 시각화 준비
        interfaces = []
        usages = []
        for line in network_data:
            parts = line.split()
            if len(parts) < 3:
                continue  # 데이터가 충분하지 않으면 다음 줄로 넘어감
            
            iface = parts[0]
            try:
                usage = float(parts[-2])  # 마지막 문자 '%' 제거 후 float로 변환
                interfaces.append(iface)
                usages.append(usage)
            except ValueError:
                continue

        # 시각화
        plt.figure(figsize=(10, 6))
        plt.bar(interfaces, usages, color='orange')
        plt.xlabel('Network Interface')
        plt.ylabel('Usage (%)')
        plt.title('Network Usage')
        plt.ylim(0, 100)
        plt.grid(True)
        plt.xticks(rotation=45)  # x축 라벨을 45도 회전하여 가로 막대 그래프로 표시
        plt.tight_layout()
        plt.show()

    except ValueError as ve:
        print(f"ValueError occurred: {ve}")
    except IndexError as ie:
        print(f"IndexError occurred: {ie}")
    except Exception as e:
        print(f"Exception occurred: {e}")

# 디스크 사용량 시각화 함수
def visualize_disk_usage(disk_data):
    try:
        # 데이터 전처리: 필요한 부분 추출
        lines = disk_data.splitlines()
        start_line = lines.index('## Disk Usage:') + 1  # 데이터 시작 라인
        disk_data = lines[start_line:]  # 필요한 데이터 부분만 추출

        # 데이터 파싱
        disks = []
        usages = []
        for line in disk_data:
            if line.startswith('/dev/'):
                parts = line.split()
                if len(parts) < 6:
                    continue
                disk = parts[0]
                try:
                    usage = float(parts[-2][:-1])  # 마지막 문자 '%' 제거 후 float로 변환
                    disks.append(disk)
                    usages.append(usage)
                except ValueError:
                    continue

        # 시각화
        plt.figure(figsize=(10, 6))
        plt.barh(disks, usages, color='lightgreen')
        plt.xlabel('Usage (%)')
        plt.ylabel('Disk')
        plt.title('Disk Usage')
        plt.xlim(0, 100)
        plt.grid(True)
        plt.tight_layout()
        plt.show()

    except ValueError as ve:
        print(f"ValueError occurred: {ve}")
    except IndexError as ie:
        print(f"IndexError occurred: {ie}")
    except Exception as e:
        print(f"Exception occurred: {e}")

# 모든 리소스 시각화 함수 호출
def visualize_all_resources(data):
    visualize_cpu_usage(data)
    visualize_network_usage(data)
    visualize_disk_usage(data)

# 예시 사용법:
if __name__ == '__main__':
    filename = 'usage_data.txt'  # 실제 파일명에 맞게 변경
    data = read_data_from_file(filename)

    visualize_all_resources(data)
```
